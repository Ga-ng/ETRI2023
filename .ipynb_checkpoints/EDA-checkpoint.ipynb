```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Load x, y, z axis data into separate arrays
x = np.loadtxt('x_data.txt')
y = np.loadtxt('y_data.txt')
z = np.loadtxt('z_data.txt')

# Define the wavelet function and scales
wavelet = 'cmor1.5-1.0'
scales = np.arange(1, 101)

# Perform continuous wavelet transform on each axis data separately
cwt_x, freqs_x = pywt.cwt(x, scales, wavelet)
cwt_y, freqs_y = pywt.cwt(y, scales, wavelet)
cwt_z, freqs_z = pywt.cwt(z, scales, wavelet)

# Calculate the power spectral density for each axis data
psd_x = (abs(cwt_x)) ** 2
psd_y = (abs(cwt_y)) ** 2
psd_z = (abs(cwt_z)) ** 2

# Define the time vector
time = np.arange(len(x))

# Create the spectrogram
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 10))

im1 = ax1.imshow(psd_x, cmap='jet', aspect='auto', extent=[time.min(), time.max(), freqs_x.min(), freqs_x.max()])
im2 = ax2.imshow(psd_y, cmap='jet', aspect='auto', extent=[time.min(), time.max(), freqs_y.min(), freqs_y.max()])
im3 = ax3.imshow(psd_z, cmap='jet', aspect='auto', extent=[time.min(), time.max(), freqs_z.min(), freqs_z.max()])

ax1.set_title('X-axis spectrogram')
ax2.set_title('Y-axis spectrogram')
ax3.set_title('Z-axis spectrogram')

ax1.set_ylabel('Frequency (Hz)')
ax2.set_ylabel('Frequency (Hz)')
ax3.set_ylabel('Frequency (Hz)')

ax3.set_xlabel('Time (s)')

fig.colorbar(im3, ax=[ax1, ax2, ax3], location='right', shrink=0.7)
plt.show()
```
