## user_label

```python
data_path = './data/dataset_2020/user30'
```

```python
user_dir_list = os.listdir(data_path)

# print(dir_list)

user_label_list = []
for element in user_dir_list:
    label_data_path = data_path + '/' + element + '/' + element + '_label.csv'
    label_data = pd.read_csv(label_data_path)
    user_label_list.append(label_data)

# print(user_label_list)
df = pd.concat(user_label_list)
df['ts'] = pd.to_datetime(df['ts'], unit='s')
df = df.sort_values('ts')

df.head()
# df.tail()
# df
```

```python
user_dir_list = os.listdir(data_path)
user_dir_list.sort()

user_label_list = []
# sensors = ['e4Acc', 'e4Bvp', 'e4Eda', 'e4Hr', 'e4Temp', 'mAcc', 'mGps', 'mGyr', 'mMag']


Acc_data_list = []
# print(user_dir_list)

acc_path = data_path + '/' + user_dir_list[0] + '/' + 'e4Acc'

acc_csv_list = os.listdir(acc_path)
acc_csv_list.sort()

# print(acc_csv_list)

df_e4Acc_fir = []
for i in range(len(acc_csv_list)):
    df_acc_fir = pd.read_csv(acc_path + '/' + acc_csv_list[i])
    df_e4Acc_fir.append(df_acc_fir)
    


df_e4Acc_fir[0]
```

```python
print(f'cnt: {df_acc_fir.count()}')
print()
print(f'min: {df_acc_fir.min()}\n')
print(f'25%: {df_acc_fir.quantile(.25)}\n')
print(f'50%: {df_acc_fir.quantile(.50)}\n')
print(f'75%: {df_acc_fir.quantile(.75)}\n')
print(f'max: {df_acc_fir.max()}\n')
print()

print(f'mean: {df_acc_fir.mean()}\n')
print(f'median: {df_acc_fir.median()}\n')
# print(f'mode: {df_acc_fir.mode()}\n')

# print(df_acc_fir.median()['x'])

instruction_list = ['df_acc_fir.min()', 'df_acc_fir.quantile(.25)', 'df_acc_fir.quantile(.50)',
                    'df_acc_fir.quantile(.75)', 'df_acc_fir.max()', 'df_acc_fir.mean()', 'df_acc_fir.median()']
acc_info = {}

for instruction in instruction_list:
    x,y,z = eval(instruction+'[\'x\']'), eval(instruction+'[\'y\']'), eval(instruction+'[\'z\']')
    
    acc_info[instruction[11:]] = [x,y,z]

    
print(acc_info)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize=(5,15))
# ax = fig.add_subplot(111, projection='3d')


i = 0
X, Y, Z = df_e4Acc_fir[i]['x'], df_e4Acc_fir[i]['y'], df_e4Acc_fir[i]['z']
# print(X)
# print(type(X))

# print(list(X))

plt.subplot(3, 1, 1)
plt.plot(df_e4Acc_fir[i]['timestamp'], list(X))

plt.subplot(3, 1, 2)
plt.plot(df_e4Acc_fir[i]['timestamp'],  list(Y))

plt.subplot(3, 1, 3)
plt.plot(df_e4Acc_fir[i]['timestamp'],  list(Z))

plt.show()


def plot_Acc(df, i):
    
    fig = plt.figure(figsize=(15,5))
    X, Y, Z = df_e4Acc_fir[i]['x'], df_e4Acc_fir[i]['y'], df_e4Acc_fir[i]['z']
#     print(X)
#     print(type(X))

#     print(list(X))

    plt.subplot(3, 1, 1)
    plt.plot(df[i]['timestamp'], list(X))

    plt.subplot(3, 1, 2)
    plt.plot(df[i]['timestamp'],  list(Y))

    plt.subplot(3, 1, 3)
    plt.plot(df[i]['timestamp'],  list(Z))

    plt.show()
```
