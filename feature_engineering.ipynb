```python
def feature_engineering(df):
    type_actions = list(df['action'].drop_duplicates())
    action_count = df['action'].value_counts()
    type_actionOptions = list(df['actionOption'].drop_duplicates())
    actionOption_count = df['actionOption'].value_counts()
    type_conditions = list(df['condition'].drop_duplicates())
    condition_count = df['condition'].value_counts()
    type_activity = list(df['activity'].drop_duplicates())
    return type_actions, action_count, type_actionOptions, actionOption_count, type_conditions, condition_count, type_activity
```

```python
# dataframe을 tensor로 변환하는 method
def dfToTensor(df, list_attributes):
    attribute_tensors = []
    for attribute in list_attributes:
        type_attributes = list(df[attribute].drop_duplicates())
        tensor = torch.Tensor([x for x in df.apply(lambda x: type_attributes.index(x[attribute]), axis=1)]).to(torch.int64)
        attribute_tensors.append(tensor)
    return attribute_tensors
```

```python
def indexToOneHot(attribute_tensors):
    return([F.one_hot(x) for x in attribute_tensors])
```
